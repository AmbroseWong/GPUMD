# NEP tutorial
## 1. Introduction
- In this tutorial, we will show how to use the `nep` executable to train a NEP potential and check the results.
- The NEP potential has been first introduced in GPUMD-v2.6 and this version wil be called NEP1. It corresponds to the following paper: 
  - Zheyong Fan, Zezhu Zeng, Cunzhi Zhang, Yanzhou Wang, Keke Song, Haikuan Dong, Yue Chen, and Tapio Ala-Nissila, Neuroevolution machine learning potentials: Combining high accuracy and low cost in atomistic simulations and application to heat transport, Phys. Rev. B **104**, 104309 (2021). https://doi.org/10.1103/PhysRevB.104.104309
- in GPUMD-v2.7 we have updated the NEP potential to NEP2, which corresponds to the following paper:
  - Zheyong Fan, Improving the accuracy of the neuroevolution machine learning potential for multi-component systems, arXiv:2109.10643 . https://arxiv.org/abs/2109.10643
- This tutorial is based on NEP2 in GPUMD-v2.7, but we will mention the differences between NEP1 and NEP2 when it is necessary.



## 2. Prerequisites
- You need to have access to an Nvidia GPU with compute capability >= 3.5. CUDA 9.0 or higher is also needed to compile the GPUMD package.
- You can download GPUMD-v2.7, unpack it, go the `src` directory, and type `make` to compile. Then, you should get three executables: `gpumd`, `phonon`, and `nep` in the `src` directory.

## 3. Train a NEP potential for PbTe
- To train a NEP potential, one must prepare two input files: `train.in` and `nep.in`. The `train.in` file contains all the training data and the `nep.in` file contains some controlling parameters.

### 3.1. Prepare the train.in file
- The `train.in` file should be prepared according to the the documentation: https://gpumd.zheyongfan.org/index.php/The_train.in_input_file
- For the example in this tutorial, we have already prepared the `trian.in` file in this folder: https://github.com/brucefan1983/GPUMD/tree/master/examples/nep/PbTe_Fan_2021/train
- This is the training data set for PbTe as used in the NEP1 paper: https://doi.org/10.1103/PhysRevB.104.104309
- It is also the training data set for PbTe as used in the NEP2 paper: https://arxiv.org/abs/2109.10643
- The only difference between NEP1 and NEP2 regarding the `train.in` file format is about the atom type: NEP1 uses the atomic number for the atom type and NEP2 uses integers from 0 to N - 1 for an N-element system. Therefore, in NEP1, Pb atoms are of type 82 and Te atoms are of type 52; in NEP2, we choose Pb atoms as type 1 and Te atoms as type 0. The user can also choose the opposite: type 0 for Pb and type 1 for Te, and the important thing is to **remember the choice**, and prepare the `xyz.in` file for the `gpumd` executable **consistently**. 
- In this example, there are only energy and force training data. In general, one can also have virial in `train.in`.
- We used the VAPS code to calculate the training data, but `nep` does not care about this. You can generate the data in `train.in` using any method that works for you. For example, before doing new DFT calculations, you can first check if there are training data publicly available and then try to convert them into the format as required by `train.in`. 

### 3.2. Prepare the `nep.in` file. 
- First, study the document about this input file: https://gpumd.zheyongfan.org/index.php/The_nep.in_input_file
- For the example in this tutorial, we have prepared the `nep.in` file in this folder: https://github.com/brucefan1983/GPUMD/tree/master/examples/nep/PbTe_Fan_2021/train
- The `nep.in` file for GPUMD-v2.7 reads:
```
num_types            2
cutoff               8 4
n_max                12 6
l_max                4
ANN                  40
reg                  0.05 0.05
population_size      50
maximum_generation   100000 
```
- We explain it line by line below:
  - There are two atoms types in the training set. This is true, as we have Pb and Te atoms.
  - The radial and angular cutoff distances are 8 angstrom and 4 angstrom, respectively.
  - The n_max parameter for the radial and angular descriptor parts are 12 and 6, respectively.
  - The l_max parameter is 4.
  - The number of neurons in the hidden layer is 40.
  - The L1 and L2 regularization parameters are both 0.05.
  - The population size is 50.
  - The maximum number of generations is 100000.
 

### 3.3. Run the `nep` executable
 - We have prepared a driver input file `input_nep.txt` in the `examples` folder. This driver input file reads:
 ```
1
examples/nep/PbTe_Fan_2021/train
```

- Now one can run the `nep` executable to train a NEP potential for PbTe. To do this, first go the directory where you can see `src`. 
  - In Linux, type this to run:
```
src/nep < examples/input_nep.txt
```
  - In Windows, type this to run:
```
src\nep < examples/input_nep.txt
```
 

## 4. Check the training results

- After running the `nep` executable, there will be some output on the screen. We encourage the user to read the screen output carefully. It can help to understand the calculation flow of `nep`. 
- Some files will be generated in the folder containing the `train.in` and `nep.in` files and will be updated every 100 generations. Therefore, one can check the results even before finishing the maximum number of generations as set in `nep.in`. 
- If the user has not studied the documentation of the output files generated by `nep`, it is time to read it here: https://gpumd.zheyongfan.org/index.php/The_output_files_for_the_nep_executable
- We will use Python to visualize the results in some of the output files next. We will need `numpy` and `matplotlib.pyplot`, so we load them first:

````python
from pylab import *
````

### 4.1. Checking the `loss.out` file. 
- We see that the $\mathcal{L}_1$ and $\mathcal{L}_2$ regularization loss functions first increase and then decreases, which indicates the effectiveness of the regularization.
- The energy loss is the root mean square error (RMSE) of energy per atom, which converges to about 0.40 meV/atom at $10^5$ generations. This is the result using NEP2 (GPUMD-v2.7). Using NEP1 (GPUMD-v2.6), the converged energy RMSE is about 0.56 meV/atom.
- The force loss is the RMSE of force components, which converges to about 37 meV/Angstrom at $10^5$ generations. This is the result using NEP2 (GPUMD-v2.7). Using NEP1 (GPUMD-v2.6), the converged force RMSE is about 50 meV/Angstrom.

````python
loss = loadtxt('loss.out')
loglog(loss[:, 0], loss[:, 1:6])
xlabel('Generation')
ylabel('Loss')
legend(['Total', 'L1-regularization', 'L2-regularization', 'Energy', 'Force'])
tight_layout()
````

### 4.2. Checking the `energy.out` file

- The dots are the raw data and the line represents the identity function used to guide the eyes.

````python
energy = loadtxt('energy.out')
plot(energy[:, 1], energy[:, 0], '.')
plot(linspace(-3.85,-3.69), linspace(-3.85,-3.69), '-')
xlabel('DFT energy (eV/atom)')
ylabel('NEP energy (eV/atom)')
tight_layout()
````

### 4.3. Checking the `force.out` file

- The dots are the raw data and the line represents the identity function used to guide the eyes.

````python
force = loadtxt('force.out')
plot(force[:, 3:6], force[:, 0:3], '.')
plot(linspace(-4,4), linspace(-4,4), '-')
xlabel('DFT force (eV/A)')
ylabel('NEP force (eV/A)')
legend(['x direction', 'y direction', 'z direction'])
tight_layout()
````

### 4.4. Checking the `virial.out` file
- In general, one can similarly check the `virial.out` file, but in this particular example, virial data were not used in the training process (no virial data exist in `train.in`). 
- However, `virial.out` still contains the predicted virial data for each structure in the training set.

````python
virial = loadtxt('virial.out')
plot(virial[:, 1], virial[:, 0], '.')
plot(linspace(-2,2), linspace(-2,2), '-')
xlabel('DFT virial (eV/atom)')
ylabel('NEP virial (eV/atom)')
tight_layout()
````

## 5. Restart and/or test
- After each 100 steps, the `nep.restart` file will be updated.
- If `nep.restart` exists and the maximum number of generations in `nep.in` is not 0, it means you want to continue the previous training. Remember not to change the parameters related to the descriptor and the number of neurons. Otherwise the restarting behavior is undefined.
- If `nep.restart` exists and the maximum number of generations in `nep.in` is 0, it means you want to make a test. In this case you will get the outputs very quickly.
- If you want to train from scratch, you need to delete `nep.restart` first (better to first make a copy of all the results from the previous training).
